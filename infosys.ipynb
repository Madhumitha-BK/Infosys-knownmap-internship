```python
x=df[['Annual Income','Spending Score']]
x
```

```python
from sklearn.preprocessing import StandardScaler
scale=StandardScaler()
scale.fit_transform(x)
```

```python
from sklearn.cluster import KMeans
```

```python
import matplotlib.pyplot as plt
```

```python
w = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=43)
    kmeans.fit(x)
    w.append(kmeans.inertia_)

plt.plot(range(1, 11), w, color='m')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia (WCSS)')
plt.title('Elbow Method')
plt.show()
```

```python
kmeans=KMeans(n_clusters=5,random_state=43)
kmeans
```

```python
df['cluster']=kmeans.fit_predict(x)
df['cluster'].value_counts().to_frame()
```

```python
plt.scatter(df['Annual Income'],df['Spending Score'], c=df['cluster'])
plt.xlabel('Annual Income',fontweight='bold',fontsize=12)
plt.ylabel('Spending Score',fontweight='bold',fontsize=12)
plt.title('Final K=5 clusters from(0-4)')
plt.show()
```
